````python
student_detail_data = {'student_id':["S1","S2","S3","S4","S5","S7","S8","S9","S11","S12"],
        'Name':["Danniella Fenton","Ryder Storey","Bryce Jensen","Ed Bernal","Kwame Morin"," Mark Smith","Bob Love","Bruce Ong","Alex Sander","Allie Chang"],
                      "cumulative_marks":[200,210,190,222,199,201,214,193,230,183]}
student_detail = pd.DataFrame(student_detail_data)
student_detail
````

### Exercise 1.1   

Include the exam score from `exam_data` for each student listed in the `student_detail` table.


### Exercise 1.2


### Exercise 1.3
Calculate the final average score, considering that the `cumulative_marks` is the sum of 3 different exams. 



````python
merged_df["average_score"]= (merged_df["cumulative_marks"]+merged_df["exam_score"])/4
merged_df
````

### Exercise 1.4
Assign a letter grade to each student based on their relative performance (percentiles): top 25% gets an A, second 25% chunk gets a B and so on. 

````python
merged_df.average_score.quantile([0,0.25,0.5,0.75,1])
````

````python
merged_df['letter_grade'] = ''
merged_df.loc[merged_df.average_score >= 63.75, 'letter_grade'] = 'A'
merged_df.loc[(59.50 <= merged_df.average_score) & (merged_df.average_score< 63.75), 'letter_grade'] = 'B'
merged_df.loc[(55.75 <= merged_df.average_score) & (merged_df.average_score < 59.50), 'letter_grade'] = "C"
merged_df.loc[(merged_df.average_score< 55.75), 'letter_grade'] = "D"
merged_df
````

print(df)
#>                         row

#### Desired Output
```

```



````python
df = pd.DataFrame(["STD, City    State",
"33, Kolkata    West Bengal",
"44, Chennai    Tamil Nadu",
"40, Hyderabad    Telengana",
"80, Bangalore    Karnataka"], columns=['row'])
print(df)
````

### Exercise 18 (Bonus)

Filter words that contain at least two vowels from the following series: 

**Input**
`ser = pd.Series(['Apple', 'Orange', 'Plan', 'Python', 'Money']) `

**Desired Output**:
```
0     Apple
1    Orange
4     Money
dtype: object
```

````python
ser = pd.Series(['Apple', 'Orange', 'Plan', 'Python', 'Money'])
ser
````

````python
from collections import Counter
result = ser.map(lambda c: sum([Counter(c.lower()).get(i, 0)
                                   for i in list('aeiou')]) >= 2)
 
print(ser[result])
````
